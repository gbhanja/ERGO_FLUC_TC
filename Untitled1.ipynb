```python
import numpy as np
import qutip as qt
import matplotlib.pyplot as plt

from tqdm import tqdm
from joblib import Parallel, delayed

from scipy.special import comb
from scipy.optimize import curve_fit
from itertools import product

############
# Parameters
############

N_arr = np.arange(2, 51, 1)

nmax = 100

ω = 1.0
ω0 = 1.0
g = 1.0

tlist = np.arange(0.01, (2*np.pi), 0.0001)

######################
# collective operators
######################

def collective_ops(N):

    Sp = qt.jmat(N/2, '+')
    Sm = Sp.dag()
    Sz = qt.jmat(N/2, 'z')

    HB = Sz + (N/2)*qt.qeye(N+1)

    return Sp, Sm, HB

############################
# Tavis-Cummings Hamiltonian
############################

def tavis_cummings(N, nmax, ω, ω0, g):

    a = qt.destroy(nmax)

    Sp, Sm, HB = collective_ops(N)

    H = (
        ω * qt.tensor(a.dag()*a, qt.qeye(N+1))
        + ω0 * qt.tensor(qt.qeye(nmax), HB)
        + g * (qt.tensor(a, Sp) + qt.tensor(a.dag(), Sm))
    )

    return H, HB

################
# Initial state
################

def initial_state(N, nmax, state):

    if state == "coherent":
        psiA = qt.coherent(nmax, np.sqrt(N))

    elif state == "fock":
        psiA = qt.basis(nmax, N)

    psiB = qt.basis(N+1, N)

    return qt.tensor(psiA, psiB)

################
# Passive state
################

def passive_state(rho, H):

    r_val, r_vec = rho.eigenstates()

    r_val = np.maximum(r_val, 0)
    r_val = r_val / np.sum(r_val)

    idx = np.argsort(r_val)[::-1]

    r_val = r_val[idx]

    e_val, e_vec = H.eigenstates()

    return np.sum(
        r_val[i] * e_vec[i] * e_vec[i].dag()
        for i in range(len(r_val))
    )

########################
# ergotropy calculation
########################

def ergotropy(rho, H):

    rho_p = passive_state(rho, H)

    return qt.expect(H, rho) - qt.expect(H, rho_p)

##############
# Pnm matrix
##############

def pnm_matrix(rho, H):

    r_val, r_vec = rho.eigenstates()

    r_val = np.maximum(r_val, 0)
    r_val = r_val / np.sum(r_val)

    idx = np.argsort(r_val)[::-1]

    r_val = r_val[idx]
    r_vec = [r_vec[i] for i in idx]

    e_val, e_vec = H.eigenstates()

    pnm = np.zeros((len(e_val), len(e_val)))

    for n in range(len(e_val)):
        for m in range(len(e_val)):

            pnm[n, m] = (
                r_val[m]
                * abs(e_vec[n].overlap(r_vec[m]))**2
            )

    return pnm, r_val, e_val

#########################################################
# Extract Dicke-sector probabilities
#########################################################

def extract_Cn_dicke(rho_b, N):

    diag = np.real(rho_b.diag())

    Cn = np.zeros(N+1)

    for n in range(N+1):

        Cn[n] = diag[n]

    Cn = Cn / np.sum(Cn)

    return Cn

#########################################################
# Probability distribution
#########################################################

def probability_distribution_energy(rho_b, N, omegaB=1.0):

    Cn_prob = extract_Cn_dicke(rho_b, N)

    deltaE = np.arange(-N, N+1) * omegaB

    P_deltaE = np.zeros(len(deltaE))

    for n in range(N+1):

        dE = (N - n) * omegaB

        idx = np.argmin(np.abs(deltaE - dE))

        P_deltaE[idx] += Cn_prob[n]

    return deltaE, P_deltaE

#########################################################
# Moments
#########################################################

def moments_from_distribution(deltaE, P_deltaE):

    mean1 = np.sum(deltaE * P_deltaE)

    mean2 = np.sum((deltaE**2) * P_deltaE)

    mean3 = np.sum((deltaE**3) * P_deltaE)

    cumulant3 = (
        mean3
        - 3*mean2*mean1
        + 2*(mean1**3)
    )

    return mean1, mean2, mean3, cumulant3

#########################################################
# Global solver options
#########################################################

opts = qt.Options(
    atol=1e-16,
    rtol=1e-14
)

#########################################################
# Compute optimal charging time
#########################################################

def compute_tau(N):

    H, HB = tavis_cummings(
        N, nmax, ω, ω0, g
    )

    HB_full = qt.tensor(
        qt.qeye(nmax),
        HB
    )

    psi0 = initial_state(
        N,
        nmax,
        "coherent"
    )

    result = qt.sesolve(
        H,
        psi0,
        tlist,
        e_ops=HB_full,
        options=opts
    )

    Eb_list = np.array(result.expect[0])

    power = Eb_list / tlist

    idx = np.argmax(power)

    τ = tlist[idx]

    return τ

#########################################################
# Parallel τ computation
#########################################################

τ_list = Parallel(n_jobs=-1)(
    delayed(compute_tau)(N)
    for N in tqdm(
        N_arr,
        desc="Computing τ"
    )
)

#########################################################
# Operator Formalism
#########################################################

def operator_formalism(i, N):

    τ = τ_list[i]

    H, HB = tavis_cummings(
        N,
        nmax,
        ω,
        ω0,
        g
    )

    psi0 = initial_state(
        N,
        nmax,
        "coherent"
    )

    result = qt.sesolve(
        H,
        psi0,
        [0, τ],
        options=opts
    )

    rho_b = result.states[-1].ptrace(1)

    E_B = qt.expect(HB, rho_b)

    r_vals, r_vecs = rho_b.eigenstates()

    r_vals = np.maximum(r_vals, 0)

    r_vals = r_vals / np.sum(r_vals)

    idx = np.argsort(r_vals)[::-1]

    r_vals = r_vals[idx]

    r_vecs = [r_vecs[i] for i in idx]

    e_vals = np.array([0] + [ω0]*N)

    E_erg = (
        E_B
        - sum(
            r_vals[j]*e_vals[j]
            for j in range(len(r_vals))
        )
    )

    W_2 = (
        qt.expect(HB**2, rho_b)
        + sum(
            r_vals[j]*(e_vals[j]**2)
            for j in range(len(r_vals))
        )
        - 2 * sum(
            e_vals[i]
            * r_vals[i]
            * qt.expect(HB, r_vecs[i])
            for i in range(len(r_vals))
        )
    )

    ΔE2 = W_2 - (E_erg)**2

    #################################################
    # Third moment
    #################################################

    term1 = qt.expect(HB**3, rho_b)

    term2 = -3 * sum(
        e_vals[j]
        * r_vals[j]
        * qt.expect(HB**2, r_vecs[j])
        for j in range(len(r_vals))
    )

    term3 = 3 * sum(
        (e_vals[j]**2)
        * r_vals[j]
        * qt.expect(HB, r_vecs[j])
        for j in range(len(r_vals))
    )

    term4 = -sum(
        r_vals[j] * (e_vals[j]**3)
        for j in range(len(r_vals))
    )

    W_3 = (
        term1
        + term2
        + term3
        + term4
    )

    mu3 = (
        W_3
        - 3 * W_2 * E_erg
        + 2 * (E_erg**3)
    )

    if ΔE2 > 1e-14:

        skewness = (
            mu3 / (ΔE2**1.5)
        )

    else:

        skewness = 0.0

    Ratio = E_erg / E_B

    return (
        N,
        τ,
        E_B,
        E_erg,
        ΔE2,
        Ratio,
        W_3,
        skewness
    )

#########################################################
# Parallel Operator Formalism
#########################################################

op_results = Parallel(n_jobs=-1)(
    delayed(operator_formalism)(i, N)
    for i, N in enumerate(
        tqdm(
            N_arr,
            desc="Operator Formalism"
        )
    )
)

#########################################################
# Work Statistics Formalism
#########################################################

def work_statistics(i, N):

    τ = τ_list[i]

    H, HB = tavis_cummings(
        N,
        nmax,
        ω,
        ω0,
        g
    )

    psi0 = initial_state(
        N,
        nmax,
        "coherent"
    )

    result = qt.sesolve(
        H,
        psi0,
        [0, τ],
        options=opts
    )

    rho_full = result.states[-1].proj()

    rho_b = rho_full.ptrace(1)

    Eb = qt.expect(HB, rho_b)

    pnm, r_val, e_val = pnm_matrix(
        rho_b,
        HB
    )

    erg = 0

    mean_E_sq = 0

    for n in range(len(e_val)):

        En = e_val[n]

        for m in range(len(e_val)):

            Em_pass = 0 if m == 0 else ω0

            ΔE = En - Em_pass

            erg += pnm[n, m] * ΔE

            mean_E_sq += (
                pnm[n, m]
                * (ΔE**2)
            )

    var = mean_E_sq - erg**2

    #################################################
    # Distribution moments
    #################################################

    deltaE, P_deltaE = (
        probability_distribution_energy(
            rho_b,
            N,
            omegaB=ω0
        )
    )

    μ1, μ2, μ3, κ3 = (
        moments_from_distribution(
            deltaE,
            P_deltaE
        )
    )

    ratio = erg / Eb

    return (
        N,
        τ,
        Eb,
        erg,
        var,
        ratio,
        μ3
    )

#########################################################
# Parallel Work Statistics
#########################################################

ws_results = Parallel(n_jobs=-1)(
    delayed(work_statistics)(i, N)
    for i, N in enumerate(
        tqdm(
            N_arr,
            desc="Work Statistics"
        )
    )
)

#########################################################
# Convert results
#########################################################

erg_OP = np.array([r[3] for r in op_results])

var_OP = np.array([r[4] for r in op_results])

third_OP = np.array([r[6] for r in op_results])

erg_WS = np.array([r[3] for r in ws_results])

var_WS = np.array([r[4] for r in ws_results])

third_WS = np.array([r[6] for r in ws_results])

#########################################################
# Plot 1
#########################################################

plt.figure(figsize=(8,5))

plt.plot(
    N_arr,
    erg_OP,
    'o-',
    label='OP Formalism'
)

plt.plot(
    N_arr,
    erg_WS,
    's--',
    label='WS Formalism'
)

plt.xlabel(r'$N$')

plt.ylabel(r'Ergotropy')

plt.title('Ergotropy vs N')

plt.legend()

plt.grid(alpha=0.4)

plt.show()

#########################################################
# Plot 2
#########################################################

plt.figure(figsize=(8,5))

plt.plot(
    N_arr,
    var_OP,
    'o-',
    label='OP Formalism'
)

plt.plot(
    N_arr,
    var_WS,
    's--',
    label='WS Formalism'
)

plt.xlabel(r'$N$')

plt.ylabel(r'Variance')

plt.title('Variance vs N')

plt.legend()

plt.grid(alpha=0.4)

plt.show()

#########################################################
# Plot 3
#########################################################

plt.figure(figsize=(8,5))

plt.plot(
    N_arr,
    third_OP,
    'o-',
    label='OP Formalism'
)

plt.plot(
    N_arr,
    third_WS,
    's--',
    label='WS Formalism'
)

plt.xlabel(r'$N$')

plt.ylabel(r'Third Moment')

plt.title('Third Moment vs N')

plt.legend()


plt.show()
```
